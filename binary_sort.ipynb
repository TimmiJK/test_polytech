```
import random

list =[]
n = 1000

for i in range (n):
  x = random.randint(0 ,1000)
  list.append(x)

list = sorted(list)

def binary_sort(array, key):
  lower_bound = 0
  upper_bound = len(array) - 1

  while lower_bound <= upper_bound:
    center = (lower_bound + upper_bound) // 2
    if array[center] == key:
      return center
      break
    elif array[center] > key:
      upper_bound = center - 1
    elif  array[center] < key:
      lower_bound = center + 1
    
  return -1 

print(list)
print(binary_sort(list, 5))
%timeit binary_sort(list, 5)
```
